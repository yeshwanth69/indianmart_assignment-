# after installation, then we import required libraries

```python
from urllib.request import Request, urlopen
from bs4 import BeautifulSoup
import csv
```

# we are requesting that url; to get access to the files of that page.

```python
URL = 'https://www.agtta.co.in/individuals.php'
req = Request(URL)
page = urlopen(req)
soup = BeautifulSoup(page, 'html.parser')
```

# by this we load the entire html file of the website.its like crtl+U

```python
print(soup)
```

```python
soup.ul #this soup construct will return all the unordered lists
```

```python
soup.title #it will return the title
```

```python
soup.body.b 
```

```python
soup.a #it will return the first link of anchor tag. 
```

```python
soup.body #returns the whole body 
```

```python
for link in soup.find_all('a'): #all the links in the file are returned
    print(link.get('href'))
```

```python
soup.section #this is the section, from where we want to get our csv file.
```

```python
individual_person_details = soup.find_all('section', attrs={'style':'margin-top: 40px;'})
```

```python
file = open('individual_files.csv', 'wt')  #creates a csv file in the folder
writer = csv.writer(file) 
```

```python
writer.writerow(['Username', 'Mobile','Firm','Address','Phone No','Route','B. Date','J. Date'])
```

```python
print (individual_person_details) #the actual data with tags.
```

```python
for individual in individual_person_details:
    name = individual.find('h3', class_="b-branches__title ui-title-inner ui-title-inner_lg").text.strip()
    unfound  = individual.find('i', class_= 'icon fa').text.strip()
    print (name + ',  ' + unfound + '  ')
   file.close()   
```
